```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#renames columns
column_names={'T1':'Temperature_in_Kitchen', 'T2': 'Temperature_in_Livingroom', 
             'T3':  'Temperature_in_Laundryroom', 'T4': 'Temperature_in_office',
              'T5': 'Temperature_in_Bathroom', 'T6': 'Temperature_outside_building',
              'T7':'Temperature_in_ironingroom', 'T8': 'Temperature_in_teenager_room',
              'T9': 'Temperature_in_Parent_room', 'RH_1':'Humidity_in_Kitchen',
              'RH_2': 'Humidity_in_Living_room', 'RH_3': 'Humidity_in_Laundary_room',
              'RH_4': 'Humidity_in_office', 'RH_5': 'Humidity_in_Bathroom',
              'RH_6': 'Humidity_outside_building', 'RH_7': 'Humidity_in_ironingroom', 
             'RH_8': 'Humidity_in_teenager_room','RH_9': 'Humidity_in_parent_room',
            'rv1': 'Random_variable_1','rv2': 'Random_variable_2', 'T_out': 'Temperature_Outside',
             'RH_out': 'Humidiity_Outside', 'Press_mm_hg': 'Pressure_mm_Hg'
             }
#command to assign new values to columns
df=df.rename(columns=column_names)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#removing dropping features date and lights
new=df.drop(columns=['lights','date'])
```

```python
new.head()
```

```python
#Using min max scaler to normalize the dataset to a common scale
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

#normalizing and droping some Features
norm=pd.DataFrame(scaler.fit_transform(new), columns=new.columns)
#dropping light
features=norm.drop(columns=['Appliances'])
Appliances=norm['Appliances']
```

```python
norm.head()
```

```python
features.head()
```

```python
#Splitting dataset into training and test dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(features,Appliances,test_size=0.3, random_state=42)
```

```python
#Linear Regresssion
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
#fit model to a training dataset
linear_model.fit(x_train, y_train)
#obtain prediction
predicted_values = linear_model.predict(x_test)
```

```python
#For Mean Average Error(MAE)
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,predicted_values)

#Residual sum of squares
rss=np.sum(np.square(y_test - predicted_values))

#Root mean square error
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test, predicted_values))
round(mae,3)

#R-squared
from sklearn.metrics import r2_score
r_squared=r2_score(y_test,predicted_values)

print('rmse','is',round(rmse,3))
print('r_squared', 'is', round(r_squared,3))
print('mae','is',round(mae,3))
print('rss','is',round(rss,3))
```

```python
#ridge regression
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=0.5)
ridge_reg.fit(x_train, y_train)
pred2 = ridge_reg.predict(x_test) 

#evaluating metrics
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,pred2)

#Residual sum of squares
rss=np.sum(np.square(y_test - pred2))

#Root mean square error
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test, pred2))
round(mae,3)

#R-squared
from sklearn.metrics import r2_score
r_squared=r2_score(y_test,pred2)

print('rmse','is',round(rmse,3))
print('r_squared', 'is', round(r_squared,3))
print('mae','is',round(mae,3))
print('rss','is',round(rss,3))
```

```python
#lasso regression
from sklearn.linear_model import Lasso
lasso_reg=Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
pred3 = lasso_reg.predict(x_test) 

#evaluating metrics
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,pred3)

#Residual sum of squares
rss=np.sum(np.square(y_test - pred3))

#Root mean square error
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test, pred3))
round(mae,3)

#R-squared
from sklearn.metrics import r2_score
r_squared=r2_score(y_test,pred3)

print('rmse','is',round(rmse,3))
print('r_squared', 'is', round(r_squared,3))
print('mae','is',round(mae,3))
print('rss','is',round(rss,3))
```
